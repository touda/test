```python
# === ハイブリッドRBFN構築コード（格子RBF + 学習済クラスタRBF + 入力・中心点可視化付き）===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
from matplotlib.patches import Ellipse
import seaborn as sns
# Google Driveと接続を行います。これを行うことで、Driveにあるデータにアクセスできるようになります。
# 下記セルを実行すると、Googleアカウントのログインを求められますのでログインしてください。
from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir('/content/drive/MyDrive/Colab Notebooks')

# === 日本語フォント設定 ===
!pip install japanize-matplotlib
import japanize_matplotlib
#matplotlib.rcParams['font.family'] = 'Hiragino Sans'
matplotlib.rcParams['axes.unicode_minus'] = False

# === データ読み込み ===
df = pd.read_csv("rainfall_2009_2023_no_year.csv")
df.columns = ['rainfall', 'soil']
df['rainfall'] = pd.to_numeric(df['rainfall'], errors='coerce')
df['soil'] = pd.to_numeric(df['soil'], errors='coerce')
df_clean = df.dropna()

# === パラメータ設定 ===
#delta_x, delta_y = 5, 1
#sigma_x, sigma_y = 145, 80
#lambda_min, lambda_max = 1, 10000

# === 標準入力によるパラメータ設定 ===
delta_x = int(input("Δx（格子のX方向間隔）を入力してください: "))         # 例: 5
delta_y = int(input("Δy（格子のY方向間隔）を入力してください: "))         # 例: 1
sigma_x = float(input("σx（X軸方向の楕円半径）を入力してください: "))    # 例: 145
sigma_y = float(input("σy（Y軸方向の楕円半径）を入力してください: "))    # 例: 80
lambda_min = float(input("λの最小値を入力してください: "))               # 例: 1
lambda_max = float(input("λの最大値を入力してください: "))               # 例: 10000

# === 格子点とクラスタ代表点抽出 ===
x_range = np.arange(0, 305, delta_x)
y_range = np.arange(0, 141, delta_y)
grid_centers, grid_counts = [], []
cluster_centers, cluster_lambdas = [], []

for cx in x_range:
    for cy in y_range:
        mask = (
            (df_clean['soil'] >= cx) & (df_clean['soil'] < cx + delta_x) &
            (df_clean['rainfall'] >= cy) & (df_clean['rainfall'] < cy + delta_y)
        )
        local_points = df_clean.loc[mask, ['soil', 'rainfall']].values
        count = len(local_points)
        if count > 0:
            # 格子中心
            grid_centers.append((cx, cy))
            grid_counts.append(count)
            # クラスタ重心（soil: X軸, rainfall: Y軸）
            kmeans = KMeans(n_clusters=1, random_state=0, n_init=10).fit(local_points)
            soil_center, rainfall_center = kmeans.cluster_centers_[0]
            cluster_centers.append((soil_center, rainfall_center))
            lam = lambda_max - (lambda_max - lambda_min) / (1 + count) #(7)式
            cluster_lambdas.append(lam)

# === RBF関数定義 ===
def rbf(x, y, center, sx, sy):
    return np.exp(-(((x - center[0])**2) / (2 * sx**2) + ((y - center[1])**2) / (2 * sy**2))) # (4)式

# === 学習：H行列と重みw ===
m = len(cluster_centers)
H = np.zeros((m, m))
for i, xi in enumerate(cluster_centers):
    for j, cj in enumerate(cluster_centers):
        H[i, j] = rbf(xi[0], xi[1], cj, sigma_x, sigma_y) #(5)式

Lambda = np.diag(cluster_lambdas)
y_teacher = np.ones(m)
w = np.linalg.inv(H.T @ H + Lambda) @ H.T @ y_teacher #(6)式

# === 応答面Z生成 ===
x_vals = np.arange(0, 296, 5)
y_vals = np.arange(0, 141, 1)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.zeros_like(X, dtype=float)

# 格子RBF（信頼度 λ を乗算）
for (cx, cy), lam in zip(grid_centers, cluster_lambdas):
    Z += lam * rbf(X, Y, (cx, cy), sigma_x, sigma_y)

# 学習済クラスタRBF（重み付き合成）
for j, cj in enumerate(cluster_centers):
    Z += w[j] * rbf(X, Y, cj, sigma_x, sigma_y)


#正規化調整
# 元のZの最小値・最大値
Z_min, Z_max = Z.min(), Z.max()

# 任意の範囲に変換したい（例：a〜b）
a = 0.0
b = 1.1

Z_norm = a + (b - a) * (Z - Z_min) / (Z_max - Z_min)

# 正規化の上下限（指定値）
#target_min = 0.00434595
#target_max = 1.07106

# 指定範囲にスケーリング
#Z_norm = target_min + (target_max - target_min) * (Z - Z_min) / (Z_max - Z_min)
print(f"Z_norm range: {Z_norm.min():.6f} ～ {Z_norm.max():.6f}")


# === 正規化 ===
#Z_min, Z_max = Z.min(), Z.max()
#Z_norm = (Z - Z_min) / (Z_max - Z_min) if Z_max != Z_min else np.zeros_like(Z)
#Z_norm = 0.0 + 1.0 * (Z - Z_min) / (Z_max - Z_min)  # → 0.1 ～ 1.1
#Z_norm = Z.copy()  # 正規化せず、Zそのまま使う
#print(f"Zの最大値 = {Z.max():.4f}, 最小値 = {Z.min():.4f}")

# === 3D出力 ===
fig1 = plt.figure(figsize=(10, 7))
ax1 = fig1.add_subplot(111, projection='3d')
surf = ax1.plot_surface(X, Y, Z_norm, cmap='plasma', edgecolor='none')
ax1.set_title("RBFN応答面（格子RBF + 学習クラスタRBF）")
ax1.set_xlabel("土壌雨量指数")
ax1.set_ylabel("1時間雨量")
ax1.set_zlabel("出力（0〜1）")
fig1.colorbar(surf, shrink=0.6, aspect=10)
plt.tight_layout()
plt.savefig("rbfn_combined_surface.png", dpi=300)
plt.close()

# === 等高線図 ===
fig2, ax2 = plt.subplots(figsize=(10, 6))
contour = ax2.contour(X, Y, Z_norm, levels=np.arange(0.0, 1.01, 0.05), colors='black', linewidths=0.5)
ax2.clabel(contour, inline=True, fontsize=8)
  # ← 線の太さ
ax2.set_title("等高線図（格子RBF + 学習クラスタRBF）")
ax2.set_xlabel("土壌雨量指数")
ax2.set_ylabel("1時間雨量")
ax2.grid(True)
plt.tight_layout()
plt.savefig("rbfn_combined_contour.png", dpi=300)
plt.close()

# === 入力・中心点の可視化 ===
fig3, ax3 = plt.subplots(figsize=(10, 6))
ax3.scatter(df_clean['soil'], df_clean['rainfall'], c='gray', s=5, label='入力データ')
ax3.scatter([pt[0] for pt in grid_centers], [pt[1] for pt in grid_centers], c='blue', s=10, label='基底関数中心（格子）')
ax3.scatter([pt[0] for pt in cluster_centers], [pt[1] for pt in cluster_centers], c='red', s=20, marker='x', label='クラスタ代表点（学習データ）')
ax3.set_title("入力データ・格子点・クラスタ代表点の分布")
ax3.set_xlabel("土壌雨量指数")
ax3.set_ylabel("1時間雨量")
ax3.legend()
ax3.grid(True)
plt.tight_layout()
plt.savefig("rbfn_input_centers_combined.png", dpi=300)
plt.close()

# === CSV保存 ===
df_csv = pd.DataFrame(np.round(Z_norm, 5), index=y_vals, columns=x_vals)
#df_csv.index.name = "1時間雨量"
#df_csv.columns.name = "土壌雨量指数"
df_csv.to_csv("rbfn_combined_output.csv", float_format="%.5f")

# === 最大出力位置の確認 ===
i_max, j_max = np.unravel_index(np.argmax(Z_norm), Z_norm.shape)
max_soil = x_vals[j_max]
max_rainfall = y_vals[i_max]
max_value = Z_norm[i_max, j_max]

print(f"✅ 最大RBFN出力位置: 土壌雨量指数 = {max_soil}, 1時間雨量 = {max_rainfall}, 出力値 = {max_value:.4f}")

# 格子内データ数の計算
grid_centers, grid_counts = [], []
for cx in x_range:
    for cy in y_range:
        mask = (
            (df_clean['soil'] >= cx) & (df_clean['soil'] < cx + delta_x) &
            (df_clean['rainfall'] >= cy) & (df_clean['rainfall'] < cy + delta_y)
        )
        count = mask.sum()
        if count > 0:
            grid_centers.append((cx, cy))
            grid_counts.append(count)

# 2Dマトリクスに格納
count_matrix = np.full((len(y_range), len(x_range)), np.nan)
for (cx, cy), count in zip(grid_centers, grid_counts):
    xi = int(cx / delta_x)
    yi = int(cy / delta_y)
    count_matrix[yi, xi] = count

# ヒートマップ作成
# 1. y_range を 20 までに限定
y_limit_index = np.where(y_range <= 10)[0]  # y=20以下の行インデックス
count_matrix_limited = count_matrix[y_limit_index, :]  # 行を制限
y_labels_limited = y_range[y_limit_index]              # ラベルも制限

# 2. ヒートマップ描画
fig4, ax4 = plt.subplots(figsize=(12, 6))
sns.heatmap(count_matrix_limited,
            xticklabels=x_range,
            yticklabels=y_labels_limited,
            cmap='YlOrRd',
            cbar_kws={'label': '格子内データ数'},
            ax=ax4)

ax4.set_title("各格子内のデータ数ヒートマップ（Y軸20まで）")
ax4.set_xlabel("土壌雨量指数 (X 軸方向格子位置)")
ax4.set_ylabel("1時間雨量 (Y 軸方向格子位置)")
plt.tight_layout()

heatmap_path = "grid_data_density_heatmap_y70.png"
plt.savefig(heatmap_path, dpi=300)
plt.close()

print("✅ ヒートマップ保存:", heatmap_path)
```
